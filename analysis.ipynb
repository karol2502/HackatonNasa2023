```python
def calculate_empty_col(tmp_df):
    empty_col = 0
    for i in range(1, 54):
        x = 0
        x = tmp_df[tmp_df.columns[i]]
        if x.isna().sum() == 180:
            empty_col += 1
    return empty_col
```

```python
train_df = pd.DataFrame()
test_df = pd.DataFrame()

x = 1
step = 180
empty_cols = 0
num_iters = 18208
for i in tqdm(range((7248 + 2502 + 2813) *step, step* num_iters, step)):
    tmp_df = df.iloc[i: i + step]
    empty_cols = calculate_empty_col(tmp_df)
    if empty_cols == 0:
        train_df = pd.concat([train_df, tmp_df])
    else:
        test_df = pd.concat([test_df, tmp_df])
```

```python
train_df.to_csv('data\\5645train.csv', index=False)
test_df.to_csv('data/5645test.csv', index=False)
```

```python
df = pd.read_csv('data\\tmp_df.csv')
target = pd.read_csv('data\\kpIndex.csv', sep=';')
target.columns = ['date', 'kp']
target['kp_test'] = target['kp'].shift(-1)

result_train = df.merge(target, on='date', how='left')
```

```python
import matplotlib.pyplot as plt
train_data = np.array([])
train_labels = np.array([])

for i in range(0, 3277440, 180):
    tmp_df = result_train.iloc[i : i+180]
    label = tmp_df.iloc[0]['kp_test']
    data = tmp_df.drop(columns=['date', 'kp', 'kp_test']).mean()
    data = np.array(data)
    train_data = np.append([train_data], data)
    train_labels = np.append([train_labels], label)
print(label)
print(train_data.shape)
```
